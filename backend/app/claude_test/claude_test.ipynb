```python
import base64
import anthropic
import os
import dotenv
```

```python
dotenv.load_dotenv()

API_KEY = os.getenv("ANTHROPIC_API_KEY")
client = anthropic.Anthropic(api_key=API_KEY)
```

```python
IMAGE_PATH = "/Users/ZapXDragon/Desktop/zap/LaTrobeUni/CAP/CV-Accessibility-of-Public-Transport/backend/app/claude_test/accessibility_feature.jpg"
MESSAGE = "Identify the main accessibility feature in this image."
```

```python
def encode_image_to_base64(image_path):
    with open(image_path, "rb") as image_file:
        encoded_string = base64.b64encode(image_file.read()).decode("utf-8")
    return encoded_string
```

```python
base64_image = encode_image_to_base64(IMAGE_PATH)
```

```python
response = client.messages.create(
    max_tokens=1000,
    messages=[
        {"role": "user",
         "content": [
             {"type": "image", "source": {
                 "type": "base64",
                 "media_type": "image/jpeg",
                 "data": base64_image
             }},
             {"type": "text", "text": MESSAGE}
             
         ]}],
    model="claude-opus-4-6"
)
```
