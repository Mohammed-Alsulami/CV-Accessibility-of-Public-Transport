```python
import anthropic
import os
import dotenv
dotenv.load_dotenv()
client = anthropic.Anthropic(api_key=os.getenv("ANTHROPIC_API_KEY"))
```

```python
message = "What is the capital of France?"

response = client.messages.create(
    max_tokens=100,
    messages=[{"role": "user", "content": message}
    ],
    model="claude-opus-4-6",
)
    
```
